```python
from matplotlib import pyplot as plt
```

```python
x = [1, 3, 5, 7, 9, 11]
y = [5, 9, 7, 3, 8, 2]
```

```python
plt.scatter(x, y, marker = '1', s = 200, color = 'black', label = "Pontos")
plt.plot(x, y, linestyle = "dashdot", color = "r", label = "Linhas")
plt.title("Gráfico com Pontos")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
plt.plot(x, y, color = "cyan", label = "Info")
plt.grid(True) #pode ser axis = 'x' ou axis = 'y' para definir a direção do grid
plt.title("Gráfico com Grid")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
plt.show()
```

```python
z = [4, 1, 2, 3]
rotulos = ["DNA", "RNA", "CITO", "ADEN"]
```

```python
plt.pie(z, labels = rotulos, autopct = '%1.1f%%')
plt.title("Gráfico de Pizza")
plt.axis("equal")
plt.show()
```

```python
plt.subplot(2, 2, 1)
plt.scatter(x, y, marker = '1', s = 200, color = 'black', label = "Pontos")
plt.plot(x, y, linestyle = "dashdot", color = "r", label = "Linhas")
plt.title("Gráfico com Pontos")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
```

```python
plt.subplot(2, 2, 2)
plt.plot(x, y, color = "cyan", label = "Info")
plt.grid(True) #pode ser axis = 'x' ou axis = 'y' para definir a direção do grid
plt.title("Gráfico com Grid")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
```

```python
plt.subplot(2, 2, 3)
plt.pie(z, labels = rotulos, autopct = '%1.1f%%')
plt.title("Gráfico de Pizza")
plt.axis("equal")
```

```python
plt.subplot(2, 2, 4)
plt.bar(x, y, color = "red", label = "Histograma")
plt.xlabel("Eixo X")
plt.ylabel("Eixo Y")
plt.legend()
```
